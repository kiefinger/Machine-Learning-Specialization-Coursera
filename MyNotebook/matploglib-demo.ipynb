```python
import math
import numpy as np
plt.axis([0,math.pi,-1.1,1.1])
t = np.arange(0,math.pi,0.1)
y1 = np.sin(math.pi*t)
y2 = np.sin(math.pi*t+math.pi/2)
y3 = np.sin(math.pi*t-math.pi/2)
plt.title('My first plot')
plt.xlabel('Counting')
plt.ylabel('Square values')
plt.text(0,0.1,'First')
plt.text(1,0.4,'Second')
plt.text(2,0.9,'Third')
plt.text(3,0.1,'Fourth')
plt.text(1.1,1,r'$y = x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.2})
plt.plot(t,y1,'b--',t,y2,'g',t,y3,'r-')
plt.grid(True)
plt.show()
```

Axis based on Date or Month

```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df = pd.DataFrame({'year': 2015,
                   'month': [1,1,2,2,3,3,4,4],
                   'day': [23,28,3,21,15,24,8,24],
                   'readings': [12,22,25,20,18,15,17,14]})
df['events'] = pd.to_datetime(df[['year','month','day']])

months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%Y-%m')
fig, ax = plt.subplots()
plt.plot(df['events'],df['readings'])
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)

            
```
